## 1.1 Чтение таблицы


```python
import pandas as pd

df = pd.read_csv("hgnc_complete_set/hgnc_complete_set.tsv", sep="\t")
df
```

## 1.2 Построение графика

```python
import matplotlib.pyplot as plt
df['locus_group'].value_counts().plot(kind='bar')

plt.title("Разные классы по столбцу locus_group")
plt.xlabel("Класс")
plt.ylabel("Количество представителей класса")
```
