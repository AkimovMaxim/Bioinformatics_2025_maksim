# Белок STING

## Краткая информация

Stimulator of interferon genes (STING) - белок, который играет важную роль во врожденном иммунитете. STING индуцирует выработку интерферонов I типа, когда клетки инфицированы внутриклеточными патогенами, такими как вирусы, микобактерии и внутриклеточные паразиты. Интерфероны I типа, опосредованные STING, защищают инфицированные клетки и близлежащие клетки от местной инфекции путем связывания с той же клеткой, которая их секретирует и соседними клетками. Он интересен для вирусологии, поскольку подавление его работы влечёт за собой ингибирование имунного ответа. Этим приёмом пользуются некоторые вирусы, таким образом, изучение метаболических путей с STING поможет невелировать эти действия со стороны патоганных агентов.
