# "Расшибешься, я тебя по частям собирать не буду," - сказал биотехнолог выращенному вирусу гриппа и отдал биоинформатику. 


# А тот взял и стал собирать...

## Семинар: SPAdes

### Код

~~~

python3 /projects/mipt_dbmp_biotechnology/soft/SPAdes-4.1.0-Linux/bin/spades.py --careful  -k 21,33,55,77,99,127 /projects/mipt_dbmp_biotechnology/genome_de_novo/7_S4_L001_R1_001.fastq -2 /projects/mipt_dbmp_biotechnology/genome_de_novo/7_S4_L001_R2_001.fastq -o ~/homeworks/hw_11/genome_assembly_results_hw/spades

~~~


### Результаты сборки

#### Контиги: 
- Всего контигов: 49.
- Контигов длиной ≥ 1000 bp: 1.
- Контигов длиной ≥ 5000 bp, 10000 bp, 25000 bp, 50000 bp: 0.
- Крупнейший контиг: 1069 bp.
- Общая длина всех контигов: 15 064 bp.

**Вывод**: Сборка крайне фрагментирована. Отсутствие длинных контигов указывает на низкое качество или неполноту сборки.

#### Метрики N50 и L50:
- N50 = 440 bp, N90 = 148 bp — низкие значения - сборки с преобладанием коротких фрагментов.
- L50 = 12, L90 = 37 - фрагментированное покрытие.

#### GC-состав и качество:
- GC = 45.47

#### Метрика auN:
- ну могло быть и получше, но геном как-то покрывает.

### По бласту:



Результаты BLAST демонстрируют высокое качество совпадения с штаммом вируса гриппа A (H3N2).

## ДЗ: SPAdes

### Код

~~~
python3 /projects/mipt_dbmp_biotechnology/soft/SPAdes-4.1.0-Linux/bin/spades.py   --careful   -k 21,33,55,77,99,127   -1 /projects/mipt_dbmp_biotechnology/genome_de_novo/7_S4_L001_R1_001.fastq   -2 /projects/mipt_dbmp_biotechnology/genome_de_novo/7_S4_L001_R2_001.fastq   -o ~/homeworks/hw_11/genome_assembly_results_hw/spades
~~~

### Почему? SPAdes лучше подходит для вирусных геномов, а увеличение параметра k отвечает большому размеру итоговой сборки. 

#### Хорошие значения метрик. Малое количество контигов. В общем всё ещё фигня какая-то, но из худшего, по метрикам - лучшее.

Чемпионат по биоинформатике. Зал 10 тысяч человек. Все сидят и жестко собирают геном вируса гриппа. Время заканчивается, команды представляют свои результаты:

Американский биоинформатик - e-value = 0,1; Query cover = 50%

Судьи головой покачали, перешли к следующему результату.

Русский биоинформатик - e-value = 0,001; Query cover = 96%

И результат хороший и покрытие. Ну вообще молодец. Подходят к китайкому биоинформатику, а тот без компа, без монитора сидит и что-то на бумажке пишет. Судьи удивляются: Как так? Биоинформатик и без компьютера. 

Собирают лист, вгоняют последовательность в BLAST:
e-value = 1e-24; Query cover = 99,9%

Рекордный результат за всю историю чемпионата!!! Судьи хватаются за голову, ничего не понимают.

Кое-как оправляются от шока, подходят к следующему участнику, а у того даже листочка нет и карандаша. Судьи спрашивают: Как же так? Что же вы делали всё это время?

А тот ему отвечает: Так мы все 9997 китайцев на один листочек старались.

**Мораль:** Легче найти 9998 китайцев, чем решить домашку по биоинформатике.
