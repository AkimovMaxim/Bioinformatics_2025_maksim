```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Сгенерим данные из рандомного распределения и посмотрим на boxplot

```python
np.random.seed(42)  
data = np.random.uniform(low=0, high=10, size=1000)

plt.figure(figsize=(8, 6))
plt.boxplot(data, vert=True, patch_artist=True)

plt.ylabel('Значения', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 2. Перейдём к заданию два: обработке предоставленных данных 

```python
data = [79, 80, 80, 80, 74, 80, 80, 79, 64, 78, 73, 78, 74, 45, 81, 48, 80, 82, 82, 70]
```

## Посмотрим квантили и медиану (тоже ж квартиль)

```python
data_sorted = sorted(data)
Q1 = np.quantile(data, 0.25)
Q2 = np.quantile(data, 0.50) 
Q3 = np.quantile(data, 0.75)

print(f"Q25: {Q1}")
print(f"Медиана: {Q2}")
print(f"Q75: {Q3}")
```

## Узнаем межквартильный размах (ΙQR)

```python
IQR = Q3 - Q1
print(f"Межквартильный размах: {IQR}")
```

## Минимум/максимум и снова медиана с квантилями

```python
min_val = min(data)
max_val = max(data)

print(f"Минимум: {min_val}")
print(f"Q25: {Q1}")
print(f"Медиана: {Q2}")
print(f"Q75: {Q3}")
print(f"Максимум: {max_val}")
```

## Рассмотрим выбросы как данные выпадающие из нижней и верхней границ (которые определяются через межквартильный размах по известной формуле)

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"Нижняя граница для выбросов: {lower_bound}")
print(f"Верхняя граница для выбросов: {upper_bound}")
print(f"Потенциальные выбросы: {outliers}")
```

## Построим boxplot

```python
plt.figure(figsize=(8, 5))
plt.boxplot(data, vert=True, patch_artist=True, widths=0.6)

plt.ylabel("Значения", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
